```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(2, 3,tight_layout = True)#領域を設定

#折れ線グラフ
y0 = [1, 1, -4, 2]
axes[0,0].plot(y0) #[0,0]は左上を示す。左上にy0をplotする

#sin関数
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)
axes[0,1].plot(x,y,linewidth = 2.0)

#ヒストグラム
xh = np.random.normal(15,5,200)
axes[0,2].hist(xh)

#散布図
x2 = np.random.normal(1,11,30)
y2 = np.random.normal(2,12,30)
axes[1,0].scatter(x2,y2)

#箱ひげ図
x3 = [10,20,30,60]
axes[1,1].boxplot(x3)

plt.show()
```
