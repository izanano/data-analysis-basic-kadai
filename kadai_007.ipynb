```python
#Step1

#ライブラリの用意
import pandas as pd
import japanize_matplotlib
import matplotlib.pyplot as plt
```

```python
#csvファイルを読み込み変数に代入する
category_df = pd.read_csv('category.csv')
sample_df = pd.read_csv('sample_pandas_6.csv')
```

```python
#中身の確認
category_df
```

```python
#中身の確認
sample_df
```

```python
#データをマージする
df = pd.merge(sample_df,category_df[['商品番号','カテゴリー']],how = 'inner',on = '商品番号')
```

```python
#マージできているか確認
df
```

```python
#Step2

#カテゴリー列の各要素数をカウント
df2 = df['カテゴリー'].value_counts()
```

```python
#カウントできているか確認
df2
```

```python
#Step2

#各要素数を棒グラフで表示
df2.plot(kind = 'bar')
plt.xlabel('カテゴリー')
plt.ylabel('個数')

#Step3

plt.show()

#ここはカテゴリー数をカウントするコードと同じセルに記述すべきか？
```

```python
#Step4

#商品番号列ごとに注文数の統計量（度数、平均など）をまとめる
df.groupby('商品番号')['注文数'].describe()
```
